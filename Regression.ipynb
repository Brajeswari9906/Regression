1. What is Simple Linear Regression?
    - Simple Linear Regression analyzes how one input variable affects one output variable by fitting a straight line to data.

         y = a + bx
         
         Where,

          x is independent variable(feature, predictor)

          y is dependent variable (target)

          a is intercept

          b is slope (rate of change)

          Date analysts use SLR for prediction, trend analysis, impact measurement, decision-making.



2. What are the key assumptions of Simple Linear Regression?
    - (a) Linearity: The relationship between the independent variable(x) and dependent variable (y) is linear.
    - (b) Independence: The observations (errors) are independent of each other.
    - (c) Homoscedasticity: The variance of the errors is constant across all value of x.
    - (d) Normality of errors: The residuals(errors) are normally distributed (important for hypothesis testing).
    - (e) No or little measurement , error in x: The independent variable is measured accurately.
    - (f) Zero mean of errors: The expected value of the error term is zero.



3. What is heteroscedasticity, and why is it important to address in regression
models?
    - Heteroscedasticity occurs in a regression model when the variance of the errors terms(residuals) is not constant across all values of the independent variable.
    - Important for (a) Biased standard errors
                    (b) Invalid hypothesis tests
                    (c) Unreliable confidence intervals
                    (d) Reduced efficiency



4. What is Multiple Linear Regression?
    - Multiple linear regression is a statistical technique used to model the relationship between one dependent variable and two or more independent variables by fitting a linear equation to observed data.



5. What is polynomial regression, and how does it differ from linear
regression?
    - Polynomial regression is a type of regression model where the relationship between the independent variable and dependent variable is modeled as a polynomial function of the independent variable.
    - It differs in such a way:

| Aspect | Linear Regression | Polynomial Regression |
|--------|------------------|----------------------|
| Relationship | Straight line | Curved (non-linear) |
| Flexibility | Low | Higher |
| Risk of overfitting | Low | Higher (for high degrees) |
| Interpretation | Simple | More complex |


6. Implement a Python program to fit a Simple Linear Regression model to
the following sample data:

      ● X = [1, 2, 3, 4, 5]

      ● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
      
      Plot the regression line over the data points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

model = LinearRegression()
model.fit(x, y)

y_pred = model.predict(x)

print("Slope (β₁):", model.coef_[0])
print("Intercept (β₀):", model.intercept_)

plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, y_pred, color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

7. Fit a Multiple Linear Regression model on this sample data:

      ● Area = [1200, 1500, 1800, 2000]

      ● Rooms = [2, 3, 3, 4]
      
      ● Price = [250000, 300000, 320000, 370000]

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Given data
Area = np.array([1200, 1500, 1800, 2000])
Rooms = np.array([2, 3, 3, 4])
Price = np.array([250000, 300000, 320000, 370000])

# Combine features
X = np.column_stack((Area, Rooms))

# Create and fit the model
model = LinearRegression()
model.fit(X, Price)

# Display results
print("Intercept:", model.intercept_)
print("Coefficient for Area:", model.coef_[0])
print("Coefficient for Rooms:", model.coef_[1])

# Predictions
predicted_prices = model.predict(X)
print("Predicted Prices:", predicted_prices)
```

9. Create a residuals plot for a regression model trained on this data:

      ● X = [10, 20, 30, 40, 50]

      ● Y = [15, 35, 40, 50, 65]
      
      Assess heteroscedasticity by examining the spread of residuals.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)
residuals = Y - Y_pred
```

Residuals Plot (code)

```python
plt.scatter(X, residuals)
plt.axhline(y=0)
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residuals Plot")
plt.show()
```

The residuals plot shows that the residuals are not evenly spread around zero for all values of X. As the value of X increases, the residuals become more spread out. This means that the variance is not constant, which indicates the presence of heteroscedasticity in the regression model.

10. Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regressionmodel.
      
      Explain the steps you would take to address these issues and ensure a robust model.

    - As a data scientist in a real estate company, predicting house prices accurately is essential. While building a regression model using features such as area, number of rooms, and location, issues like heteroscedasticity and multicollinearity may arise, which can affect the reliability of the model.

      Heteroscedasticity occurs when the variance of residuals is not constant across all levels of the independent variables. It can be detected using residual plots. To address this issue, transformations like logarithmic transformation of house prices can be applied to stabilize variance. Alternatively, Weighted Least Squares or robust standard errors can be used to obtain reliable coefficient estimates.

      Multicollinearity occurs when two or more independent variables are highly correlated with each other. It is commonly detected using a correlation matrix or Variance Inflation Factor (VIF). High multicollinearity makes coefficient estimates unstable and difficult to interpret. This issue can be handled by removing or combining highly correlated variables, or by applying regularization techniques such as Ridge or Lasso regression. Principal Component Analysis (PCA) can also be used to convert correlated variables into uncorrelated components.

      After applying these corrective measures, the model should be retrained and validated. Residual plots and VIF values should be checked again to ensure that the issues have been reduced. Model performance can be evaluated using metrics such as R² and RMSE.
